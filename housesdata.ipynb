```python
import pandas as pd
```

```python
data=pd.read_csv("D:\luxury house\Luxury_Housing_Bangalore.csv")
```

```python
data
```

```python
data['Micro_Market']=data['Micro_Market'].str.title().str.strip()
data
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data['Ticket_Price_Cr']=data['Ticket_Price_Cr'].astype(str).str.replace(' Cr','', regex=False).str.replace('₹','',regex=False).astype(float)
```

```python
data.isnull().sum()
```

```python
data['Amenity_Score']=data.groupby('Micro_Market')['Amenity_Score'].transform(lambda x: x.fillna(x.mean()))
```

```python

data['Configuration']=data['Configuration'].astype(str).str.lower()
```

```python
data['NRI_Buyer']=data['NRI_Buyer'].map({'yes':True, 'no':False})
```

```python

data['Buyer_Comments']=data['Buyer_Comments'].fillna('no commands')
```

```python
data=data[(data['Ticket_Price_Cr']>0.5) & (data['Ticket_Price_Cr']<60)]
```

```python
data=data[data['Unit_Size_Sqft']>0]
```

```python
data['Price_per_sqft']= (data['Ticket_Price_Cr']*1e7)/data['Unit_Size_Sqft']
```

```python
data['Transaction_Type'].value_counts()
```

```python
data.info()
```

```python
data['Purchase_Quarter']=pd.to_datetime(data['Purchase_Quarter'])
```

```python
conn_mysql = mysql.connector.connect(
    host="localhost",
    user="root",
    password="root"
)
cursor_mysql = conn_mysql.cursor()
print("MySQL connection established!")
```

```python
cursor_mysql.execute("CREATE DATABASE IF NOT EXISTS house;")
print("MySQL database 'house' created successfully!")

cursor_mysql.execute("USE house;")  # Select database
cursor_mysql.execute("""
    CREATE TABLE IF NOT EXISTS house_data (
        Property_ID VARCHAR(255),
        Micro_Market VARCHAR(255),
        Project_Name VARCHAR(255),
        Developer_Name VARCHAR(255),
        Unit_Size_Sqft float,
        Configuration VARCHAR(255),
        Ticket_Price_Cr float,
        Transaction_Type VARCHAR(255),
        Buyer_Type VARCHAR(255),
        Purchase_Quarter Date,
        Connectivity_Score float,
        Amenity_Score float,
        Possession_Status VARCHAR(255),
        Sales_Channel VARCHAR(255),
        NRI_Buyer BOOLEAN,
        Locality_Infra_Score float,
        Avg_Traffic_Time_Min int,
        Buyer_Comments VARCHAR(255),
        Price_per_sqft float                                       
                                              
    );
""")
conn_mysql.commit()
print("Table 'house' created successfully in MySQL!")
```

```python
data_list = data.values.tolist()
query = """
    INSERT INTO house_data( Property_ID,
        Micro_Market,
        Project_Name,
        Developer_Name,
        Unit_Size_Sqft,
        Configuration,
        Ticket_Price_Cr,
        Transaction_Type,
        Buyer_Type,
        Purchase_Quarter,
        Connectivity_Score,
        Amenity_Score,
        Possession_Status,
        Sales_Channel,
        NRI_Buyer,
        Locality_Infra_Score,
        Avg_Traffic_Time_Min,
        Buyer_Comments,
        Price_per_sqft )
    VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s);
"""
cursor_mysql.executemany(query, data_list)
conn_mysql.commit()
print("Data inserted using to_list()")
```
